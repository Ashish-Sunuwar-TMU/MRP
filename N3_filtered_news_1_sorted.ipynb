```python
import pandas as pd

# load the filtered file, parsing Date
df = pd.read_csv(
    "filtered_news_1.csv",
    parse_dates=["Date"],
    dtype={
        "Article_title":   str,
        "Stock_symbol":    str,
        "Textrank_summary":str
    },
    low_memory=False
)

# sort by Stock_symbol, then by Date
df.sort_values(by=["Stock_symbol", "Date"], inplace=True)

# write out the sorted file
df.to_csv("filtered_news_1_sorted.csv", index=False)
```
