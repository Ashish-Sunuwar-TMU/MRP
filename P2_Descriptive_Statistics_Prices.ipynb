```python
# imports & define file path
import pandas as pd
file_path = '/content/drive/MyDrive/MRP/initial_merged_prices.csv'
```

```python
# load the merged prices
df = pd.read_csv(file_path, parse_dates=['date'])
```

```python
# numeric descriptive statistics
num_cols = ['volume','open','high','low','close','adj close']
numeric_stats = df[num_cols].describe()          # count, mean, std, min, 25%, 50%, 75%, max
print("Numeric Descriptive Statistics")
display(numeric_stats)
```

```python
# missing-value percentages
missing_pct = df.isnull().mean() * 100
print("\n Missing Value % per Column ")
display(missing_pct.to_frame('missing_pct'))
```

```python
# symbol counts & distribution
symbol_counts = df['symbol'].value_counts()
symbol_summary = symbol_counts.describe()        # count, unique, top, freq, mean rows/symbol, etc.
top10 = symbol_counts.head(10)
bot10 = symbol_counts.tail(10)
print("\n Symbol-Row Counts Summary ")
display(symbol_summary.to_frame('rows_per_symbol').T)
print("Top 10 symbols by row-count:")
display(top10.to_frame('count'))
print("Bottom 10 symbols by row-count:")
display(bot10.to_frame('count'))
```

```python
# date range and per-year trends
min_date, max_date = df['date'].min(), df['date'].max()
print(f"\nDate range: {min_date.date()} → {max_date.date()}")
df['year'] = df['date'].dt.year
yearly_counts = df['year'].value_counts().sort_index()
print("\n Rows per Year ")
display(yearly_counts.to_frame('count'))
```
